```python
import numpy as np
from sklearn.model_selection import train_test_split
import os
import matplotlib.pyplot as plt
```

```python
# Define categories and data containers
categories = [
    'DU_LICH', 'GIAI_TRI', 'GIAO_DUC', 'KHOA_HOC', 'KINH_DOANH', 'PHAP_LUAT', 'SUC_KHOE', 'THE_THAO'
]

data = [[] for _ in categories]

# Function to read and process data from a file
def process_data(category, filename):
    with open(f'./raw_vnexpress_titles/raw_data/{filename}.txt') as file:
        for line in file:
            if line.strip():
                data[category].append(line.strip())

# Populate data and labels
for category, name in enumerate(categories):
    process_data(category, name)

labels = [list(np.ones(len(data[i]), dtype=int) * i) for i in range(len(categories))]
```
