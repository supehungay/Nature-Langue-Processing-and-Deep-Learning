```python
import underthesea
import numpy as np
# from nltk.tokenize import word_tokenize
import re
import gensim.downloader as api
```

```python
with open('../lecture1/text.txt') as f:
    txt = f.read().strip().replace("\n", " ")
f.close()
txt
```

```python
txt_lower = txt.lower()
txt_lower
```

```python
txt_token = re.split("\W+", txt_lower)
txt_token
```

```python
text_processing = txt_token
# Stopword removal
with open('vietnamese-stopwords.txt') as st:
    stopword = st.read().strip().splitlines();

sw_in_txt = []

for text in text_processing:
    if text in stopword:
        text_processing.remove(text)
        sw_in_txt.append(text)
 
```
